**Problem 3**

```
def fact(n, factorial=1):
  if n==0:
    return 1
  else:
    for i in range(1,n+1):
      factorial=factorial*i
    return factorial

fact(7)
```

**Problem 5**

```
def g(x):
  value = (x**4) - (2*(x**3)) - (17*(x**2)) + (4*x) + 30 
  return value

def g_prime(x):
  value = (4*(x**3)) - (6*(x**2)) - (34*x) + 4
  return value

g_prime(-2)

```

**Problem 6**

```
def newtons_method(starting_guess,n):
  x_j = starting_guess
  for i in range(1,n+1):
    x_j = x_j - (g(x_j)/g_prime(x_j))
    
  return x_j

newtons_method(10,5)
```

**Problem 7**

```
def integration(m):
  E = 1 - (1/np.exp(1))
  for j in range(1, m+1):
    E = 1 - j*E
  return E

integration(5)
```
